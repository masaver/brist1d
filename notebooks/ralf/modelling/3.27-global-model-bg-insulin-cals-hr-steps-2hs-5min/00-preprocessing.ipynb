# Preprocessing

* New imputation strategy
    * Interpolate max 2 consecutive missing values
    * bfill and ffill max 1 missing value
    * median for the rest      
* Use data and additional data 2h to train the model


```python
from data import load_data

train_data, test_data = load_data()

unique_patients = test_data['p_num'].unique()
train_data = train_data[train_data['p_num'].isin(unique_patients)]
test_data = test_data[test_data['p_num'].isin(unique_patients)]

train_data.head()
```

```python
from pipelines import preprocessing_pipeline, standardization_pipeline

train_data = preprocessing_pipeline.fit_transform(train_data)
test_data = preprocessing_pipeline.transform(test_data)
```
